```python
import tweepy
import csv
from credentials.twitter_credential import *
from credentials.google_credential import *
from google.cloud import language_v1
from google.cloud.language_v1 import enums
```

```python
auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
client = language_v1.LanguageServiceClient()
type_ = enums.Document.Type.PLAIN_TEXT
language = "en"
```

```python
f = open('data/company_list.txt')
for line in f:
    line = line.strip('\n')
    csvCursor = open('data/' + line + '.csv', 'a')
    
    csvWriter = csv.writer(csvCursor)

    for tweet in tweepy.Cursor(api.search, q=line, count=100000, lang="en", since="2000-01-01", include_entities=True).items():
            preprocessed_tweets = tweet.text.replace('\n','').replace('\t','').lower().encode('ascii', 'ignore')
            print(tweet.text.encode('utf-8'))
            if(len(preprocessed_tweets.split()) > 20):
                print(str(tweet.text.encode('utf-8')))
                try:
                    document = {"content": preprocessed_tweets, "type": type_, "language": language}
                    response = client.classify_text(document)
                    for category in response.categories:
                        if(category.name.startswith("/News") or category.name.startswith("/Finance")):
                            csvWriter.writerow([str(tweet.created_at), str(tweet.id_str), str(tweet.text.encode('utf-8')), 
                                                str(tweet.user.id), str(tweet.user.name), str(tweet.user.screen_name), 
                                                str(tweet.user.location), str(tweet.user.url), str(tweet.user.description), 
                                                str(tweet.place), str(tweet.entities.hashtags), str(tweet.entities.urls), 
                                                str(tweet.entities.user_mentions)])  

                except:
                    print("~~~~~~~~~~~~~~~~~~~~~~~exception~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                    
    csvCursor.close()
```
